# D. Filtering Data#

```python
# menampilkan daftar lagu yang populer di indonesia
data[ data['country'] == 'indonesia' ]
```

```python
# menampilkan daftar lagu diatas tahun 2015
data[data['year'] > 2015]
```

```python
# Menampilkan daftar lagu yang populer di indonesia diatas 2015

data[ (data['country'] == 'indonesia') &
     (data['year'] > 2015)
     ]
```

```python
# Menampilkan daftar lagu yang populer di indonesia diatas 2015 dan top genre adalah pop
# urutkan berdasarkan durasi terbesar ke terkecil
data[ (data['country'] == 'indonesia') &
     (data['year'] > 2015) &
     (data['top genre'] == 'indonesian pop')
     ].sort_values('dur', ascending=False)
```

# E. Group by

```python
# Mencari rata-rata durasi lagu per country
# urutkan dari kecil ke besar

data_duration = data.groupby('country')['dur'].mean().sort_values()
data_duration
```

```python
# Mengubah ke dataframe
data_duration = data_duration.to_frame().reset_index()
data_duration.columns = ['country', 'dur']
data_duration
```

```python
# Plot duration

data_duration.plot(x= 'country',
                   y= 'dur',
                   title= 'tingkat durasi (rata-rata)',
                   figsize= (15,5), #ini ukuran canvas untuk chartnya
                   kind='bar')
```

# F. Save dataset

```python
# save data_duration

data_duration.to_csv('data_duration.csv', index=False) 
#index false agar index nya tidak ikut tersimpan
#jika index True maka index akan ikut tersimpan dan akan muncul sebagai kolom unnamed
```
